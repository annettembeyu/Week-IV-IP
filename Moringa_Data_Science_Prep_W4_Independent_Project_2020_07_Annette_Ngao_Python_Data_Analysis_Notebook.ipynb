## Analysis 

#### Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.

#### checking the data type for the variables

```python
auto_1.info()
```

#### converting the year,month,day,hour,minute to string

```python
#auto_1[["year","month","day","hour", "minute"]].astype("object") 
#df[["a", "b"]] = df[["a", "b"]].apply(pd.to_numeric)
auto_1.year = auto_1.year.astype(str)
auto_1.month = auto_1.month.astype(str)
auto_1.day = auto_1.day.astype(str)
auto_1.hour = auto_1.hour.astype(str)
auto_1.minute = auto_1.minute.astype(str)
```

#### Confirming that the data type for the five columns has been changed

```python
auto_1.info()
```

#### Converting the year,month,day,hour and minute into one 

```python
auto_1['date'] = auto_1['year'] + ':' + auto_1['month'] + ':' + auto_1['day'] + ':' + auto_1['hour'] + ':' + auto_1['minute']
#df['date'] = df['year'] + '-' +nex
```

#### Displaying the data after converting

```python
auto_1
```

```python
sorted_auto =auto_1.sort_values(by=['date'], ascending = False)
sorted_auto
```

```python
sorted_auto['Electric_usage_in_Bluecar'] = sorted_auto['Bluecar_counter'].diff()
sorted_auto
```

```python
[(sorted_auto.City == "Paris") & (sorted_auto.Electric_usage_in_Bluecar < 0)].groupby('hour')['hour'].count().head(1)
```
